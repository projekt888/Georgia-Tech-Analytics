```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merge_table = pd.merge(bitcoin_df, dash_df, on="Date")
merge_table.head(10)
```

```python
# Rename columns so that they are differentiated
merge_table.columns = ['Date', 'Open Bitcoin', 'High Bitcoin', 'Low Bitcoin', 'Close Bitcoin', 'Volume Bitcoin',
              'Market Cap Bitcoin', 'Open Dash', 'High Dash', 'Low Dash', 'Close Dash', 'Volume Dash',
              'Market Cap Dash'] 


merge_table.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
```

```python
# Collecting best open for Bitcoin and Dash
openf = []
closef = []
volf = []
for x in merge_table.columns:
    if (x.startswith("Open")):
        print (x)
        print (merge_table[x].max())
        openf.append(merge_table[x].max())


# Collecting best close for Bitcoin and Dash
for x in merge_table.columns:
    if (x.startswith("Close")):
        print (x)
        print (merge_table[x].max())
        closef.append(merge_table[x].max())


# Collecting the total volume for Bitcoin and Dash
for x in merge_table.columns:
    if (x.startswith("Vol")):
        print (x)
        print (merge_table[x].max())
        volf.append(merge_table[x].max())
```

```python
# Creating a summary DataFrame using above values

d = {'Open': openf, 'Close': closef, 'Vol': volf}
df = pd.DataFrame(d, index = ['bitcoin','dash'])
df.head()
```
