```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
print("The minimum is " + str(ted_df["views"].min()))
print("The maximum is " + str(ted_df["views"].max()))
```

```python
import math
# Create bins in which to place values based upon TED Talk views

# Create labels for these bins

minbb = 50000
maxbb = 48000000

step = ((maxbb-minbb) / 5)
bins = []
names = []

for i in range(0,6):
    bins.append(math.trunc(minbb + (i * step)))

for i in range (0,5):
    names.append(str(i) + " STARS")

print(bins)
print(names)
print(len(bins))
print(len(names))
```

```python
# Slice the data and place it into bins
ted_df["View Score Summary"] = pd.cut(ted_df["views"], bins, labels=names)
ted_df.head()
```

```python
# Place the data series into a new column inside of the DataFrame

# Creating a group based off of the bins
ted_dff = ted_df.groupby("View Score Summary")
ted_dff.head()
```
