```python
# Dependencies
import pandas as pd
```

```python
# Load in File from resources
# 'movie_scores.csv'
file = "Resources/movie_scores.csv"
```

```python
# Read and display with pandas
df_original = pd.read_csv(file)
df_original.head()
```

```python
# List all the columns the table provides
print(df_original.columns)
columns = list(df_original.columns)
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB
# Reorganizing the columns using double brackets
new_cols = ["FILM"]

for c in columns:
    if(c[0:4] == "IMDB"):
        new_cols.append(c)

organized_df = df_original[new_cols]
organized_df.head()
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
only_billys = organized_df.loc[organized_df["IMDB"] >= 7, :]
only_billys
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
#ER - ask about logical and
only_billys = organized_df.loc[(organized_df["IMDB"] >= 7) & (organized_df["IMDB_user_vote_count"] < 20000), :]
only_billys
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.

file = "Resources/output_scores.csv"

only_billys.to_csv(file)
```
