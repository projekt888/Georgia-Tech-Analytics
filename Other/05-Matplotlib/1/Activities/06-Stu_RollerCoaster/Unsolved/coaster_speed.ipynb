```python
%matplotlib notebook
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
Danger_Drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
print(len(RailGun))


x_axis = np.arange(0, 130, 10)
plt.plot(x_axis, Danger_Drop,color = 'red',label="Danger Drop")
plt.plot(x_axis, RailGun,color = 'green',label="Rail Gun")


plt.title('Roller Coaster',fontsize=20)
plt.grid(True)

plt.xlabel('Coaster Runtime',fontsize=12)

plt.ylabel('Speed',fontsize=12)
plt.ylim(5, 92)
plt.xlim(0, 120)

plt.legend(loc="best")

plt.show()
```
