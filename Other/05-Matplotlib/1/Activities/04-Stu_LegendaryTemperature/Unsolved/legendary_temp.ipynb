```python
import numpy as np
import matplotlib.pyplot as plt

def f_to_c(in_f):
    return ((in_f - 32) * 0.56)

njtempf = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

njtempc = np.array([f_to_c(xi) for xi in njtempf])
#plt.plot(df.index, df[drugs[i]], marker = markers[i],color=colors[i], label=drugs[i] )
x_axis = np.arange(1, 13, 1)
plt.plot(x_axis, njtempf, marker='+',color='blue',label="Fahrenheit")
plt.plot(x_axis, njtempc,color='#CF0920',marker='s',label="Celsius")


plt.ylabel('Degrees')
plt.xlabel('Month')
plt.legend(loc="best")
plt.title('Temp in NJ')
#plt.figure(figsize=(20,20))


plt.show()
```
